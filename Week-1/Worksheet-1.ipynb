### Part D

```python
def spiral_order(arr):
    row = len(arr)
    col = len(arr[0])
    ans = []
    top ,bottom ,left ,right = 0, row, 0, col
    while top < bottom and left < right : 

        # top row
        for i in range (left,right):
            ans.append(arr[top][i])
        top += 1

        # right column
        for i in range (top,bottom):
            ans.append(arr[i][right - 1])
        right -= 1

        # bottom row
        if top < bottom :
            for i in range (right - 1, left - 1, -1):
                ans.append(arr[bottom - 1][i])
            bottom -= 1

        #left row
        if left < right :
            for i in range (bottom - 1, top - 1, -1):
                ans.append(arr[i][left])
            left += 1
            
    return ans
```

```python
spiral_list = spiral_order(random_arr.tolist())
spiral_list
print(spiral_list)
```

## Question 2

```python
import pandas as pd
```

```python
df = pd.read_csv('./videogamesales.csv')
df
```

### Part A

```python
df['Global_Sales'] = df.iloc[:,-4:].sum(axis = 1)
df = df.sort_values(by='Global_Sales', ascending = False)
df
```

### Part B

```python
import matplotlib.pyplot as plt
```

```python
genre_sales = df.groupby('Genre')['Global_Sales'].sum()
genre_sales
```

```python
genre_sales.plot(kind='bar', figsize=(10,5))
plt.ylabel('Total Global Sales')
plt.show()
```
